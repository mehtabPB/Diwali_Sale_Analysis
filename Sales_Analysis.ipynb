```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("filepath.csv",encoding='latin1')
print(data)
```

```python
## check rows and columns
data.shape
```

```python
#check first 5 and last 5 rows
data.head()
```

```python
data.tail()
```

```python
# check information for Data cleaning 
data.info()
```

```python
# Drop the column Status and unnamed as they are empty
data.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
data.info()
```

```python
# check for all null values
pd.isnull(data).sum()
```

```python
# Drop the null values 
data.dropna(inplace=True)
data.shape

#inplace is a method to save the action after performing it , like droping a row and then save the dataset
```

```python
pd.isnull(data).sum()
```

```python
#change the data type of 'Amount' from float to int
data.columns
data['Amount']=data['Amount'].astype('int')
data['Amount'].dtypes
```

```python
#check for all the given columns
data.columns
```

```python
#change the column name from 'Amount ' to Price
data.rename(columns={'Amount':'Price'},inplace=True)
```

```python
# describe() method returns description in data in the dataFrame (mean , count , std 25% etc )
data.describe()
```

```python
# EDA : EXPLORATORY DATA ANALYSIS
data.columns
```

# Gender

```python
graph=sns.countplot(x='Gender',data=data)
# put values above the graph
for bars in graph.containers:
    graph.bar_label(bars)
```

```python
# check the total spends by male and females:
salesbygender=data.groupby(['Gender'],as_index=False)['Price'].sum().sort_values(by='Price',ascending=False)
print(salesbygender)
```

```python
graph=plt.bar(salesbygender['Gender'],salesbygender['Price'])
for i,x in enumerate(salesbygender['Price']):
    plt.text(i,x,str(x),ha='center',va='bottom')
plt.show(x)

```

### CONCLUSION : FROM ABOVE GRAPH WE CAN SEE THAT MOST OF THE BUYERS ARE FEMALES AND EVEN THE PURCHASING POWER OF FEMALE IS GREATER THAN MEN

# AGE

```python
# Graph for purchasing the age group also divided with the Gender
g=sns.countplot(x='Age Group',data=data,hue='Gender')
for bar in g.containers:
    g.bar_label(bar)
```

## Most of the buyers are Females aged between 26-35 

# STATE

```python
data.columns
```

```python
# Total number orders from the top 10 states
sales_state=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
for i,x in enumerate(sales_state['Orders']):
    plt.text(i,x,str(x),ha='center',va='bottom')
```

```python
# Total amount spended by each state 
sale_amount=data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,8)})
sns.barplot(x='State',y='Amount',data=sale_amount)
for i,x in enumerate(sale_amount['Amount']):
    plt.text(i,x,str(x),ha='center',va='bottom')
```

## From the above graph we can see that unexpectedly most of the orders are from UP, Maharashtra and Karnataka respectively but total most amount is spent from UP, Karnataka then Maharashtra.

# Marital Status

```python
ans=sns.countplot(x='Marital_Status',data=data)
sns.set(rc={'figure.figsize':(5,5)})
for x in ans.containers:
    ans.bar_label(x)
    
```

```python
sale_Gender=data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sale_Gender,hue='Gender')
```

##  From the above graph we can see that most of the buyers are married females and also have the highest purchasing power

# OCCUPATION


```python
data.columns
```

```python
ocu=sns.countplot(x='Occupation',data=data)
sns.set(rc={'figure.figsize':(18,8)})
for x in ocu.containers:
    ocu.bar_label(x)
```

```python
data['Marital_Status']=data['Marital_Status'].astype(str)
sale_ocu=data.groupby(['Occupation','Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sale_ocu,hue='Marital_Status')
sns.set(rc={'figure.figsize':(18,8)})
```

```python
sale_ocu=data.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sale_ocu,hue='Gender')
sns.set(rc={'figure.figsize':(18,8)})
```

## From the above graph we can say that the three most purchasing sectors are IT and Aviation followed by Healthcare sector

# PRODUCT CATEGORY

```python
sns.countplot(x='Product_Category',data=data)
sns.set(rc={'figure.figsize':(25,8)})
```

```python
# total amount spent to buy product category
sale_product=data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=sale_product)
sns.set(rc={'figure.figsize':(25,8)})
```

## From the above graph we can see that most sold products are clothing but most amount is spent on food category
